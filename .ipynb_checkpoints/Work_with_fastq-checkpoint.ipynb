```python
def read_fastq(filename):
    sequences = []
    qualities = []
    with open(filename, "r") as fh:
        while True:
            fh.readline()
            seq = fh.readline().rstrip()
            fh.readline()
            Q_encoded = fh.readline().rstrip()
            if len(seq) == 0:
                break
            sequences.append(seq)
            qualities.append(Q_encoded)
    return sequences, qualities
```

```python
seqs, quals = read_fastq("SRR835775_1.first1000.fastq")
```

```python
# Helper function to convert into Phred + 33
def phred33_to_Q(char):
    return ord(char) - 33
```

```python
def collect_quality_each(qualities : list) -> tuple:
    """
    Q, quantities
    """
    qual_dict = {}
    for qual_string in qualities:
        for qual in qual_string:
            base_Q = phred33_to_Q(qual)
            qual_dict[base_Q] = qual_dict[base_Q] + 1 if base_Q in qual_dict.keys() else 0
    sorted_qual_dict = dict(sorted(qual_dict.items()))
    return sorted_qual_dict.keys(), sorted_qual_dict.values()
```

```python
Q, quantities = collect_quality_each(quals)
```

```python
# Draw a histogram of base quality
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.bar(Q, quantities)
plt.show()
```

```python
# Read by positions
import numpy
```

```python
def find_gc_by_position(multiple_sample_sequences : list) -> list:
    """
    Assume that every sample has the same # of base pairs - simply take the length of first sample
    """
    len_base_pairs = len(multiple_sample_sequences[0])
    gc_array = numpy.array([0 for _ in range(len_base_pairs)])
    total_array = gc_array.copy()
    for i in range(len_base_pairs):
        for sample in multiple_sample_sequences:
            gc_array[i] += 1 if sample[i] in ["G", "C"] else 0
            total_array[i] += 1
    gc_array = [gc_array[i] / float(total_array[i]) for i in range(len(gc_array)) if total_array[i] != 0]
    return gc_array
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
len_base_pairs = len(seqs[0])
plt.plot(range(len_base_pairs), find_gc_by_position(seqs))
plt.xlabel("Base pair position")
plt.ylabel("GC Concentration")
plt.show()
```
